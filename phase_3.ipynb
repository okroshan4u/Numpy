```python
import numpy as np
```

```python
#Data Structure [restaurent_id, 2021,2022,2023,2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("===== Zomato sells data =====")
print("\n Slawes data shapes ",sales_data.shape)

# print("\n Sample data for 1st 3 restaurent: \n",sales_data[0:3])
# print("\n Sample data for 1st 3 restaurent: \n",sales_data[:3])

print("\n Sample data for first 3 restaurent: \n",sales_data[:, 2:]) # so the first is row and then second is coloumn
# here sales_data[:,2:] --> here first part : means give all the rows and 2: means give all the columns from 2 to the last








```

```python
print(np.sum(sales_data,axis=0))
```

```python
# avoid the first col that is the id

print("Total sales of the company restaurent wise",np.sum(sales_data[:,1:],axis =1))

yearly_total = np.sum(sales_data[:,1:],axis = 0)
print("Yearly sales of the restaurent",yearly_total)
```

```python
# Minimum sales per restaurent 
min_sales = np.min(sales_data[:,1:],axis=1)
print(min_sales)
```

```python
#Maximum sales per year

max_sales = np.max(sales_data[:,1:],axis = 0)
print(max_sales)
```

```python
# Average sales per restaurent

avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
#cummulative sum per year
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cummulative sales across all restaurent")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# vector operation

vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition",vector1 + vector2)
print("Vector multiplication",vector1*vector2)

print("Dot product",np.dot(vector1,vector2))
```

```python
angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```

```python
# operations on each element of the matrix

restaurent_type = np.array(["biryani","chinese","pizza","burger","cafe"])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper",vectorized_upper(restaurent_type))


```

```python
# Brodcasting in numpy array

monthly_avg = sales_data[:,1:] /12
print(monthly_avg)
```
